```python
import os
import getpass
from dotenv import load_dotenv

# Load environment variables from .env
load_dotenv()

# Fetch Gemini API key
api_key = os.environ.get("GEMINI_API_KEY") or os.environ.get("GOOGLE_API_KEY")
if not api_key:
    api_key = getpass.getpass("Enter your Gemini API key: ")

os.environ["GOOGLE_API_KEY"] = api_key
os.environ["GEMINI_API_KEY"] = api_key
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import SystemMessage, HumanMessage

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

from langchain_core.prompts import ChatPromptTemplate
prompts_tempelate =  ChatPromptTemplate.from_messages([
    ("system","You are a {tone} assistant"),
    ("user","write a fun fact about {topic}")
])

usertopic = input("ask on your topics ? ")
usertone = input("set  your tone   ")


usersprompt = prompts_tempelate.invoke({"topic":usertopic,"tone":usertone})
llm.invoke(usersprompt).content

```
